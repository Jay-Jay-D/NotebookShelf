```python
import numpy as np
import pandas as pd

import iisignature as isig

import seaborn as sns
sns.set_style("darkgrid")

%matplotlib inline
```

```python
dim = 4
path=np.random.uniform(size=(391,dim))
signature=isig.sig(path,2)
s=isig.prepare(dim,2)
logsignature=isig.logsig(path,s)
logsignature
```
